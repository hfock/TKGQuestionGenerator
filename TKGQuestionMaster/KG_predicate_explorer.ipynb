```python
import pandas as pd
import TKGQuestionMaster.Generator as qm
```

```python
def get_an_entry_for_each_predicate(df, predicate_set):
    new_df = pd.DataFrame()
    for predicate in predicate_set:
        sliced_df = df.loc[df['predicate'] == predicate]
        sliced_df = sliced_df.iloc[0:1, :]
        new_df = pd.concat([new_df, sliced_df])
    return new_df
```

```python
df1 = pd.read_csv('../data/processed_datasets/TKG_readable_WIKIDATA12k.csv', index_col=0)
df1 = df1[(df1['until'] != '####')]
df1 = df1[(df1['from'] != '####')]
wikidata12k_predicates = set(df1['predicate'].tolist())
get_an_entry_for_each_predicate(df1, wikidata12k_predicates).to_csv('showcase_data/wikidata12k.csv')
```

```python
def icews_split_and_remove_predicate(string):
    splitted = string.split("',")
    if len(splitted) > 1:
        string = splitted[0]
        string = string[2: len(string)]
    else:
        string = splitted[0]
        string = string[2: len(string)-2]
    return string

df2 = pd.read_csv('../data/processed_datasets/TKG_readable_icews14.csv', index_col=0)
df2['predicate'] = df2['predicate'].apply(lambda x: icews_split_and_remove_predicate(x))

icews14_predicates = set(df2['predicate'].tolist())
get_an_entry_for_each_predicate(df2, icews14_predicates).to_csv('showcase_data/icews14.csv')
```

```python
df3 = pd.read_csv('../data/processed_datasets/TKG_readable_icews05-15.csv', index_col=0)
df3['predicate'] = df3['predicate'].apply(lambda x: icews_split_and_remove_predicate(x))

icews05_15_predicates = set(df3['predicate'].tolist())
get_an_entry_for_each_predicate(df3, icews05_15_predicates).to_csv('showcase_data/icews05-15.csv')
```

```python
df4 = pd.read_csv('../data/processed_datasets/TKG_readable_YAGO11k.csv', index_col=0)
df4['from'] = df4['from'].apply(lambda x: x[:4])
df4['until'] = df4['until'].apply(lambda x: x[:4])
df4 = df4[df4['from'] != '####']
df4 = df4[df4['until'] != '####']
# df4['predicate'] = df4['predicate'].apply(lambda x: icews_split_and_remove_predicate(x))
#
YAGO11k_predicates = set(df4['predicate'].tolist())
get_an_entry_for_each_predicate(df4, YAGO11k_predicates).to_csv('showcase_data/YAGO11k.csv')
```

```python
df5 = pd.read_csv('../data/processed_datasets/TKG_readable_CronKGQA.csv', index_col=0)
CronKGQA_predicates = set(df5['predicate'].tolist())
get_an_entry_for_each_predicate(df5, CronKGQA_predicates).to_csv('showcase_data/CronKGQA.csv')
```
